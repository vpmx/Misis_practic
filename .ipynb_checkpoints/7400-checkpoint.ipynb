```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../7400/22- 16 март.xlsx')
```

```python
df.head()
```

```python
df = df.drop(df.index[0])
df.head(2)
```

```python
df['Дата и Время'] = pd.to_datetime(df['Дата и Время'], dayfirst=True)
df.head(3)
```

```python
df["Дата и Время"] = pd.to_datetime(df["Дата и Время"],format='%m/%d/%Y %H:%M')
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Глубина'] = df['Глубина'].astype(float)
df['Вес на крюке'] = df['Вес на крюке'].astype(float)
df['Высота крюка'] = df['Высота крюка'].astype(float)
df['Скорость СПО'] = df['Скорость СПО'].astype(float)
df['Нагр.на дол'] = df['Нагр.на дол'].astype(float)
df['Момент на роторе'] = df['Момент на роторе'].astype(float)

df['Глубина долота'] = df['Глубина долота'].astype(float)
df['Давление'] = df['Давление'].astype(float)
df['Мех.скорость средняя'] = df['Мех.скорость средняя'].astype(float)
df['Ходы насоса 1'] = df['Ходы насоса 1'].astype(float)
df['Ходы насоса 2'] = df['Ходы насоса 2'].astype(float)
df['Расход вх.'] = df['Расход вх.'].astype(float)

df['Темп.вх'] = df['Темп.вх'].astype(float)
df['Темп.вых'] = df['Темп.вых'].astype(float)
df['Плотность вх'] = df['Плотность вх'].astype(float)
df['Плотность вых.'] = df['Плотность вых.'].astype(float)
df['Поток на вых.'] = df['Поток на вых.'].astype(float)
df['Рабочая емк.1'] = df['Рабочая емк.1'].astype(float)

df['Рабочая емк.2'] = df['Рабочая емк.2'].astype(float)
df['БПР 1'] = df['БПР 1'].astype(float)
df['БПР 2'] = df['БПР 2'].astype(float)
df['Долив'] = df['Долив'].astype(float)
df['ЦСГО'] = df['ЦСГО'].astype(float)
df['ЦСГО 2'] = df['ЦСГО 2'].astype(float)

df['Сумм. актив ёмк.'] = df['Сумм. актив ёмк.'].astype(float)
df['С1'] = df['С1'].astype(float)
df['С2'] = df['С2'].astype(float)
df['С3'] = df['С3'].astype(float)
df['iC4'] = df['iC4'].astype(float)
df['nC4'] = df['nC4'].astype(float)
df['iC5'] = df['iC5'].astype(float)
df['nC5'] = df['nC5'].astype(float)
df['Сумм.газ.хромат'] = df['Сумм.газ.хромат'].astype(float)
```

```python
df.isnull().sum()
```

```python
df.set_index('Дата и Время', inplace=True)
```

```python
print(df.index.min())
print(df.index.max())
```

```python
df= df.replace(-999.25, np.nan)
```

```python
df.head()
```

```python
df = df.resample('1min').mean()
df.shape
```

```python
df.isna().sum()
```

```python
С1=np.nanmedian(df['С1'])
С2=np.nanmedian(df['С2'])
С3=np.nanmedian(df['С3'])
iC4=np.nanmedian(df['iC4'])
nC4=np.nanmedian(df['nC4'])
iC5=np.nanmedian(df['iC5'])
nC5=np.nanmedian(df['nC5'])
sum_med=np.nanmedian(df['Сумм.газ.хромат'])

df['С1']=df['С1'].fillna(С1)
df['С2']=df['С2'].fillna(С2)
df['С3']=df['С3'].fillna(С3)
df['iC4']=df['iC4'].fillna(iC4)
df['nC4']=df['nC4'].fillna(nC4)
df['iC5']=df['iC5'].fillna(iC5)
df['nC5']=df['nC5'].fillna(nC5)
df['Сумм.газ.хромат']=df['Сумм.газ.хромат'].fillna(sum_med)
```

```python
df['Глубина'] = df['Глубина'].fillna(method='ffill')
df['Вес на крюке'] = df['Вес на крюке'].fillna(method='ffill')
df['Высота крюка'] = df['Высота крюка'].fillna(method='ffill')
df['Скорость СПО'] = df['Скорость СПО'].fillna(method='ffill')
df['Нагр.на дол'] = df['Нагр.на дол'].fillna(method='ffill')
df['Момент на роторе'] = df['Момент на роторе'].fillna(method='ffill')
df['Глубина долота'] = df['Глубина долота'].fillna(method='ffill')
df['Обороты'] = df['Обороты'].fillna(method='ffill')
df['Давление'] = df['Давление'].fillna(method='ffill')
df['Мех.скорость средняя'] = df['Мех.скорость средняя'].fillna(method='ffill')
df['Ходы насоса 1'] = df['Ходы насоса 1'].fillna(method='ffill')
df['Ходы насоса 2'] = df['Ходы насоса 2'].fillna(method='ffill')
df['Расход вх.'] = df['Расход вх.'].fillna(method='ffill')
df['Темп.вх'] = df['Темп.вх'].fillna(method='ffill')
df['Темп.вых'] = df['Темп.вых'].fillna(method='ffill')
df['Плотность вх'] = df['Плотность вх'].fillna(method='ffill')
df['Плотность вых.'] = df['Плотность вых.'].fillna(method='ffill')
df['Поток на вых.'] = df['Поток на вых.'].fillna(method='ffill')
df['Рабочая емк.1'] = df['Рабочая емк.1'].fillna(method='ffill')
df['Рабочая емк.2'] = df['Рабочая емк.2'].fillna(method='ffill')
df['БПР 1'] = df['БПР 1'].fillna(method='ffill')
df['БПР 2'] = df['БПР 2'].fillna(method='ffill')
df['Долив'] = df['Долив'].fillna(method='ffill')
df['ЦСГО'] = df['ЦСГО'].fillna(method='ffill')
df['ЦСГО 2'] = df['ЦСГО 2'].fillna(method='ffill')
df['Сумм. актив ёмк.'] = df['Сумм. актив ёмк.'].fillna(method='ffill')
```
