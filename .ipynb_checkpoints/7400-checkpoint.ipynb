```python
rotation = ndf['Обороты'].resample('1min').apply(lambda x : max(x, key = abs))
pump1 = ndf['Ходы насоса 1'].resample('1min').apply(lambda x : max(x, key = abs))
pump2 = ndf['Ходы насоса 2'].resample('1min').apply(lambda x : max(x, key = abs))
```

```python
ndf.drop(['Обороты','Ходы насоса 1', 'Ходы насоса 2'],axis=1,inplace=True)
ndf.head()
```

```python
# Агррегируем по 1мин и берем среднее значение

ndf_mean = ndf.resample('1min').mean()
ndf_mean.shape
```

```python
# Агррегируем по 1мин и берем макс значение

ndf_max = ndf.resample('1min').max()
ndf_max.shape
```

```python
ndf_mean['Обороты'] = rotation.values
ndf_mean['Ходы насоса 1'] = pump1.values
ndf_mean['Ходы насоса 2'] = pump2.values
```

```python
ndf_max['Обороты'] = rotation.values
ndf_max['Ходы насоса 1'] = pump1.values
ndf_max['Ходы насоса 2'] = pump2.values
```

```python
С1=np.nanmedian(ndf_mean['С1'])
С2=np.nanmedian(ndf_mean['С2'])
С3=np.nanmedian(ndf_mean['С3'])
iC4=np.nanmedian(ndf_mean['iC4'])
nC4=np.nanmedian(ndf_mean['nC4'])
iC5=np.nanmedian(ndf_mean['iC5'])
nC5=np.nanmedian(ndf_mean['nC5'])
sum_med=np.nanmedian(ndf_mean['Сумм.газ.хромат'])
BPR1=np.nanmedian(ndf_mean['БПР 1'])
BPR2=np.nanmedian(ndf_mean['БПР 2'])
SCGO=np.nanmedian(ndf_mean['ЦСГО'])

ndf_mean['С1']=ndf_mean['С1'].fillna(С1)
ndf_mean['С2']=ndf_mean['С2'].fillna(С2)
ndf_mean['С3']=ndf_mean['С3'].fillna(С3)
ndf_mean['iC4']=ndf_mean['iC4'].fillna(iC4)
ndf_mean['nC4']=ndf_mean['nC4'].fillna(nC4)
ndf_mean['iC5']=ndf_mean['iC5'].fillna(iC5)
ndf_mean['nC5']=ndf_mean['nC5'].fillna(nC5)
ndf_mean['Сумм.газ.хромат']=ndf_mean['Сумм.газ.хромат'].fillna(sum_med)
ndf_mean['БПР 1']=ndf_mean['БПР 1'].fillna(BPR1)
ndf_mean['БПР 2']=ndf_mean['БПР 2'].fillna(BPR2)
ndf_mean['ЦСГО']=ndf_mean['ЦСГО'].fillna(SCGO)
```

```python
ndf_max['С1']=ndf_max['С1'].fillna(С1)
ndf_max['С2']=ndf_max['С2'].fillna(С2)
ndf_max['С3']=ndf_max['С3'].fillna(С3)
ndf_max['iC4']=ndf_max['iC4'].fillna(iC4)
ndf_max['nC4']=ndf_max['nC4'].fillna(nC4)
ndf_max['iC5']=ndf_max['iC5'].fillna(iC5)
ndf_max['nC5']=ndf_max['nC5'].fillna(nC5)
ndf_max['Сумм.газ.хромат']=ndf_max['Сумм.газ.хромат'].fillna(sum_med)
ndf_max['БПР 1']=ndf_max['БПР 1'].fillna(BPR1)
ndf_max['БПР 2']=ndf_max['БПР 2'].fillna(BPR2)
ndf_max['ЦСГО']=ndf_max['ЦСГО'].fillna(SCGO)
```
